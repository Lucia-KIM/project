```python
# 주중 주말 합쳐서 하나의 데이터프레임 만들기
everyday = pd.concat([weekday, weekend], ignore_index=True )
everyday.sort_values(by=['일자'], axis=0, inplace=True, ignore_index=True)
raining_day = everyday[['일자', '강수여부']]
```
