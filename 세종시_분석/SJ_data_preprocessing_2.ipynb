```python
import pandas as pd
week = pd.read_csv('./3/아파트실거래_년_월_일.csv',encoding='cp949')


# 데이터를 저장할 list(리스트)를 만듭니다. 
weekly = [] # 열에 추가할 각 행을 For로 순회합니다, 

for row in week['계약일']: # 이 값보다 크면, 
    if row > 21:
        weekly.append('4주차') # 아니고, 이 값보다 크면, 
    elif row > 14:
        weekly.append('3주차') # 아니고, 이 값보다 크면, 
    elif row > 7: 
        weekly.append('2주차') # 아니고, 이 값보다 크면, 
    else: 
        weekly.append('1주차') 
        
week['계약주'] = weekly

```

```python
#쿼터별 
import pandas as pd
qua = pd.read_csv('./3/아파트실거래_년_월_일.csv',encoding='cp949')


# 데이터를 저장할 list(리스트)를 만듭니다. 
quater = [] # 열에 추가할 각 행을 For로 순회합니다, 

for row in qua['계약월']: # 이 값보다 크면, 
    if 1>=row<=3:
        quater.append('1') # 아니고, 이 값보다 크면, 
    elif 4>=row<=6:
        quater.append('2') # 아니고, 이 값보다 크면, 
    elif 6>=row<=9: 
        quater.append('3') # 아니고, 이 값보다 크면, 
    else: 
        quater.append('4') 
        
qua['계약분기'] = quater
```

```python
qua1=qua.groupby(['계약분기'])['거래금액'].mean()
```

```python
plt.figure(figsize=(10, 5))
ax = qua1.plot()
```

```python
qua1.head()
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,10))
# use unstack()
ax=qua.groupby(['계약분기','구분'])['거래금액'].mean().unstack().plot(ax=ax)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
#계절별 
import pandas as pd
sea = pd.read_csv('./3/아파트실거래_년_월_일.csv',encoding='cp949')


# 데이터를 저장할 list(리스트)를 만듭니다. 
seasons = [] # 열에 추가할 각 행을 For로 순회합니다, 

for row in sea['계약월']: # 이 값보다 크면, 
    if 3>=row<=5:
        seasons.append('봄') # 아니고, 이 값보다 크면, 
    elif 6>=row<=8:
        seasons.append('여름') # 아니고, 이 값보다 크면, 
    elif 9>=row<=11: 
        seasons.append('가을') # 아니고, 이 값보다 크면, 
    else: 
        seasons.append('겨울') 
        
sea['계절'] = seasons
```

```python
import pandas as pd
size = pd.read_csv('./3/아파트실거래_년_월_일.csv',encoding='cp949',engine='python')
```

```python
size.head()
```

```python
size['전용면적'].describe()
```

```python
size1 = [] # 열에 추가할 각 행을 For로 순회합니다, 

for row in size['전용면적']: # 이 값보다 크면, 
    if row<40:
        size1.append('1') # 아니고, 이 값보다 크면, 
    elif 40>=row<51.4:
        size1.append('2') # 아니고, 이 값보다 크면, 
    elif 51.4>=row<62.8: 
        size1.append('3') # 아니고, 이 값보다 크면, 
    elif 62.8>=row<79.35: 
        size1.append('4') # 아니고, 이 값보다 크면
    elif 79.35>=row<95.9: 
        size1.append('5') # 아니고, 이 값보다 크면,
    elif 95.9>=row<110: 
        size1.append('6')
    elif 110>=row<135: 
        size1.append('7')
    else :
        size1.append('8')
        
size['전용면적'] = size1
```

```python
size2=size.groupby(['전용면적'])['거래금액'].mean()
```

```python
plt.figure(figsize=(10, 5))
ax = size2.plot()
```
