# 1. 서울시 기상데이터 살펴보기 

```python
weather_df = pd.read_csv('./서울시_기상데이터.csv', encoding='cp949')
weather_df
```

```python
# NaN 0으로 처리
weather_df = weather_df.fillna(0)
weather_df
```

```python
# 일시 컬럼을 일자와 시간으로 분리 
weather_df['일자'] = 0
for i in range(len(weather_df)):
    weather_df['일자'][i] = weather_df['일시'].iloc[i][:10]
weather_df
```

```python
weather_df['시간'] = 0
for i in range(len(weather_df)):
    weather_df['시간'][i] = weather_df['일시'].iloc[i][11:]
weather_df
```

```python
# 필요한 컬럼만 남기고 정린
weather_df = weather_df[['일시','일자','시간', '기온','강수량','적설량']]
weather_df
```

## 1-1. 강수 여부에 따른 구분
#### 눈/비가오면 사람들이 자차를 많이 이용할 것이다라는 가정하에 교통량이 많은 시간보다 1시간 앞서 비가오는지 확인

- 서울시 교통량 조사에 따르면 주중에는 오전 7시-9시, 오후 6시~7시에 교통량이 많음
- 주말에는 13시-14시, 16~17시 사이 교통량이 많음
- 이 시간대를 기준으로 정리

### 주말에 눈/비가오면? 

```python
import datetime as dt

# datetime 형식으로 자료형 바꿔주기
weather_df['일자'] = pd.to_datetime(weather_df['일자'])

# 요일 컬럼 넣어주기 (월요일:0 ~ 일요일:6)
# 주말(토,일)은 5,6
weather_df['요일'] = weather_df['일자'].dt.weekday
weather_df
```

## 1) 평일 눈/비가오는지 여부에 따른 파생변수 만들기

```python
# 평일
weekday = weather_df[(weather_df['요일'] != 5) & (weather_df['요일'] != 6)]
weekday
```

```python
# 주중에는 오전 7시, 오후 5시로 선택

# weekday = weekday[(weekday['시간']=='7:00') | (weekday['시간']=='17:00')]
# weekday
```

```python
# 주중에는 오전 7시로 선택 
weekday = weekday[(weekday['시간']=='7:00')]
weekday
```

```python
weekday.iloc[0]['강수량']
```

```python
weekday['강수여부'] = 0

for i in range(len(weekday)):
    if weekday.iloc[i]['강수량'] != 0.0:
        weekday.iloc[i]['강수여부'] = 1
    if weekday.iloc[i]['적설량'] != 0.0:
        weekday.iloc[i]['강수여부'] = 1
weekday
```

## 2) 주말 눈/비가오는지 여부에 따른 파생변수 만들기

```python
# 주말
weekend = weather_df[(weather_df['요일'] == 5) | (weather_df['요일'] == 6)]
weekend
```

```python
# 주말에는 12시 기준으로 선택

weekend = weekend[(weekend['시간']=='12:00')]
weekend
```

```python
weekend['강수여부'] = 0

for i in range(len(weekend)):
    if weekend.iloc[i]['강수량'] != 0.0:
        weekend.iloc[i]['강수여부'] = 1
    if weekend.iloc[i]['적설량'] != 0.0:
        weekend.iloc[i]['강수여부'] = 1
weekend
```

```python
# 주중 주말 합쳐서 하나의 데이터프레임 만들기
everyday = pd.concat([weekday, weekend], ignore_index=True )
everyday.sort_values(by=['일자'], axis=0, inplace=True, ignore_index=True)
everyday
```

```python
raining_day = everyday[['일자', '강수여부']]
raining_day
```
